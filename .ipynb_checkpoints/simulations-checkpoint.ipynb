```python
from uutils import *
import numpy as np
from math import *
import sympy
import random
from lemke import *
import seaborn as sns
from matplotlib import pyplot as plt
import timeit
```

### number of iterations and size of the interaction matrix

```python
n = 100
nbr_iter = 50
```

### Symmetric case, interactions = $I - \frac{W}{\alpha \sqrt{n}}$

### Phase transition phenomenon (on p-property) for $\alpha = 2$

```python
alphas = np.linspace(0.5,3,25)

alpha_det_proba = {}
alpha_p_matrix_proba = {}

for alpha in alphas:
    det_matrix = []
    p_matrix_result = []
    cannot_solve = 0
    for _ in range(nbr_iter):
        normalized_wigner = get_wigner((n,n)) / (alpha * sqrt(n))
        matrix = np.eye(n) - normalized_wigner
        
        det_matrix.append(np.linalg.det(matrix))

        pm, J = p_matrix_Rohn(matrix)
        if pm == 1:
            is_p_matrix = True
            p_matrix_result.append(is_p_matrix)
        elif pm == -1:
            cannot_solve += 1
        else: 
            is_p_matrix = False
            p_matrix_result.append(is_p_matrix)
    
    alpha_det_proba[alpha] = np.sum(np.array(det_matrix)>0) / (nbr_iter - cannot_solve)
    alpha_p_matrix_proba[alpha] = np.sum( np.array(p_matrix_result).astype(int) ) / (nbr_iter - cannot_solve)
    
print(r"W symmetric" "\n")
print(f"Montecarlo over {nbr_iter} estimations,\nmatrix shape {n}x{n}\n")
print(f"Rohn algorithm, cannot solve the p-matrix problem: {cannot_solve} times \n\n")

plt.plot(alpha_det_proba.keys(), alpha_det_proba.values(), alpha = 0.4, color = "grey")
plt.axvline(x = 2, linestyle = "-", color = "darkorange")
plt.title(r"Probability of $\det > 0$ for $I - \frac{W}{\alpha \sqrt{n}}$ according to $\alpha$", fontsize=15)
plt.xlabel(r"$\alpha$", fontsize=15)
plt.show();

plt.plot(alpha_p_matrix_proba.keys(), alpha_p_matrix_proba.values(), alpha = 0.4, color = "grey")
plt.axvline(x = 2, linestyle = "-", color = "darkorange")
plt.title(r"Probability of p-property of $I - \frac{W}{\alpha \sqrt{n}}$ according to $\alpha$", fontsize=15)
plt.xlabel(r"$\alpha$", fontsize=15)
plt.show();
```

### Non-symmetric case, interactions = $I - \frac{\Gamma}{\alpha \sqrt{n}}$

### Phase transition phenomenon (on p-property) for $\alpha = 1$

```python
alphas = np.linspace(0.5,10,30)

alpha_det_proba = {}
alpha_p_matrix_proba = {}

for alpha in alphas:
    det_matrix = []
    p_matrix_result = []
    cannot_solve = 0
    for _ in range(nbr_iter):
        normalized_matrix = np.random.normal(loc=0.0, scale=1.0, size=(n,n)) / (alpha * sqrt(n))
        matrix = np.eye(n) - normalized_matrix
        
        det_matrix.append(np.linalg.det(matrix))
        
        pm, J = p_matrix_Rohn(matrix)
        if pm == 1:
            is_p_matrix = True
            p_matrix_result.append(is_p_matrix)
        elif pm == -1:
            cannot_solve += 1
        else: 
            is_p_matrix = False
            p_matrix_result.append(is_p_matrix)
    
    alpha_det_proba[alpha] = np.sum(np.array(det_matrix)>0) / (nbr_iter - cannot_solve)
    alpha_p_matrix_proba[alpha] = np.sum( np.array(p_matrix_result).astype(int) ) / (nbr_iter - cannot_solve)
    
print(r"$\Gamma$ non symmetric" "\n")
print(f"Montecarlo over {nbr_iter} estimations,\nmatrix shape {n}x{n}\n")
print(f"Rohn algorithm, cannot solve the p-matrix problem: {cannot_solve} times \n\n")

plt.plot(alpha_det_proba.keys(), alpha_det_proba.values(), alpha = 0.4, color = "grey")
plt.axvline(x = 1, linestyle = "-", color = "darkorange")
plt.title(r"Probability of $\det > 0$ for $I - \frac{\Gamma}{\alpha \sqrt{n}}$ according to $\alpha$", fontsize=15)
plt.xlabel(r"$\alpha$", fontsize=15)
plt.show();

plt.plot(alpha_p_matrix_proba.keys(), alpha_p_matrix_proba.values(), alpha = 0.4, color = "grey")
plt.axvline(x = 1, linestyle = "-", color = "darkorange")
plt.title(r"Probability of p-property of $I - \frac{\Gamma}{\alpha \sqrt{n}}$ according to $\alpha$", fontsize=15)
plt.xlabel(r"$\alpha$", fontsize=15)
plt.show();
```
